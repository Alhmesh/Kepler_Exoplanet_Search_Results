**Forth step**: now we need to seprate the numerical and categorical data.

```python
import numpy as np

numerical = kepler.select_dtypes(include=['integer', 'float']).columns
print('The numerical parameters (integers), (float):\n ', numerical)
categorical = kepler.select_dtypes(include=['object']).columns
print('The categorical parameters :\n', categorical)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.hist(kepler['ra'], bins = 15, color='lightblue', edgecolor='black')
plt.xlabel('RA')
plt.ylabel('Freq')
plt.title('RA Distibution')
plt.show()
```
